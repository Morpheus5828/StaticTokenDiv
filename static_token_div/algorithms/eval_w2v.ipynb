```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
embedding_file = "embeddings.txt"
embeddings = {}
```

```python
with open(embedding_file, "r") as f:
    vocab_size, embedding_dim = map(int, f.readline().split())
    for line in f:
        parts = line.strip().split()
        word = parts[0]
        vector = np.array([float(x) for x in parts[1:]])
        embeddings[word] = vector

eval_file = "Le_comte_de_Monte_Cristo.100.sim.txt"
eval_data = pd.read_csv(eval_file, sep=" ", header=None, names=["m", "m_plus", "m_minus"])
```

```python
def cosine_similarity_between(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
success_count = 0
```

```python
def evaluate_triplet(m, m_plus, m_minus):
    if m not in embeddings or m_plus not in embeddings or m_minus not in embeddings:
        return False
    sim_m_plus = cosine_similarity(embeddings[m], embeddings[m_plus])
    sim_m_minus = cosine_similarity(embeddings[m], embeddings[m_minus])
    return sim_m_plus > sim_m_minus

for _, row in eval_data.iterrows():
    if evaluate_triplet(row["m"], row["m_plus"], row["m_minus"]):
        success_count += 1
```

```python
accuracy = success_count / len(eval_data)
print(f"Taux de réussite : {accuracy * 100:.2f}%")
```
